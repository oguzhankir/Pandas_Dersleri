#### <font color='pink'>Concatenate </font> 

```python
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 1, 2, 3],
)
df1
```

```python
df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },
    index=[4, 5, 6, 7],
)

df2
```

```python
df3 = pd.DataFrame(
    {
        "A": ["A8", "A9", "A10", "A11"],
        "B": ["B8", "B9", "B10", "B11"],
        "C": ["C8", "C9", "C10", "C11"],
        "D": ["D8", "D9", "D10", "D11"],
    },
    index=[8, 9, 10, 11],
)
df3
```

```python
dataframes = [df1, df2, df3] #dataframe listesi
dataframes
```

- **Default olarak axis = 0**

```python
sonuc = pd.concat(dataframes, axis = 0) # birleştirelim
print(df1, "\n", df2, "\n", df3)
print(sonuc)
```

- **axis = 1 olarak değiştirdiğimizde ne oluyor görelim**

```python
sonuc = pd.concat(dataframes, axis = 1)
print(df1 ,"\n", df2 ,"\n", df3)
sonuc
```

- **ignore_index i inceleyelim**

```python
#bunun için ilk olarak indexleri sıralı halden çıkarıyoruz
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 1, 2, 5],
)

df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },
    index=[0, 1, 7, 3],
)


df3 = pd.DataFrame(
    {
        "A": ["A8", "A9", "A10", "A11"],
        "B": ["B8", "B9", "B10", "B11"],
        "C": ["C8", "C9", "C10", "C11"],
        "D": ["D8", "D9", "D10", "D11"],
    },
    index=[0, 1, 8, 3],
)
print(df1 ,"\n", df2 ,"\n", df3)
```

```python
dataframes = [df1, df2, df3] #dataframe listesi
sonuc = pd.concat(dataframes, ignore_index = False, axis = 0) # birleştirelim #default olarak ignore_index = false
print(df1, "\n", df2, "\n", df3)
print(sonuc)
```

```python
sonuc = pd.concat(dataframes, ignore_index=True)
sonuc
```

- **Keys parametresini inceleyelim**

```python
sonuc = pd.concat(dataframes, keys = ["df1", "df2", "df3"], ignore_index= False) # keys parametresi gireceksek igrnore_index false olmalı
sonuc
```

- Görüldüğü üzere bir multiindex yapısı oluştu. Şimdi **<font color='pink'>"key"</font> lerden birini loc ile çağıralım**

- **DataFrame ile serileri de birleştirebilirsiniz.**

```python
s1 = pd.Series(["X0", "X1", "X2", "X3"], name="X")
print(s1)
print(df1)
```

```python
sonuc = pd.concat([df1, s1], axis=1, ignore_index= False)
print(sonuc)
```

- **axis = 0 ile deneyelim**

```python
sonuc = pd.concat([df1, s1], axis=0)
print(sonuc)
```

```python
sonuc = pd.concat([df1, s1], axis=0, ignore_index=True)
print(df1)
print(s1)
print(sonuc)
```

- **DataFrame ile birden fazla seriyi birleştirmek**

```python
s2 = pd.Series(["0", "1", "2", "3"])

sonuc = pd.concat([df1, s2, s2, s2], axis=1, ignore_index= False)
print(s2)
print(df1)
print(sonuc)
```

- **birden fazla seriyi birleştirmek**

```python
s3 = pd.Series([0, 1, 2, 3])

s4 = pd.Series([0, 1, 2, 3])

s5 = pd.Series([0, 1, 4, 5])

sonuc1 = pd.concat([s3, s4, s5], axis=1)


print(sonuc1)
```

- **bir de keys parametresi vererek deneyelim**

```python
sonuc2 = pd.concat([s3, s4, s5], axis=1, keys=["red", "blue", "yellow"])
print(sonuc2)
```

- **iki adet dataframe oluşturalım ve merge parametrelerinden dolayı bunları left ve right değişkenlerine atayalım**

```python
left = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)


right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)

print(left)
print(right)
```

- **bu iki dataframe i <font color='pink'>"*key*"</font> kolonundan birleştirmek için <font color='pink'>"*on*"</font> parametresine key yazalım** 

```python
sonuc = pd.merge(left, right, on="key")
print(left)
print(right)
print(sonuc)

```

#### <font color='pink'>**Cross join**</font>

- İlk olarak mantığı anlamak için küçük dataframe ler üzerinden işlem yapalım.

```python
df1 = pd.DataFrame({"left":["A","B"]})
df2 = pd.DataFrame({"right":[1,2]})

print(df1)
print(df2)
```

Görüldüğü üzere crossjoin veri setleri arasındaki bütün kombinasyonları bir araya getirir

```python
pd.merge(df1, df2, how = "cross")
```

- Şimdi eski verisetlerimizi kullanalım ve cross_join yaptığımızda veri setinin ne kadar büyüdüğünü gözlemleyelim.

```python
cross_join = pd.merge(su_elektrik, dogalgaz, how = "cross", indicator=True)
```

```python
cross_join.head(50)
```

```python
cross_join.shape
```

- **cross join kullandığımızda bütün kombinasyonları yaptığı için *"on"* parametresi kullanılamaz.**

```python
pd.merge(su_elektrik, dogalgaz, on = "Order ID", how = "cross", indicator=True) 
```

```python
cross_join._merge.value_counts()
```

- Birkaç join işlemi de concat ile yapalım

```python
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 1, 2, 3],
)

df4 = pd.DataFrame(
    {
        "B": ["B2", "B3", "B6", "B7"],
        "D": ["D2", "D3", "D6", "D7"],
        "F": ["F2", "F3", "F6", "F7"],
    },
    index=[2, 3, 6, 7],
)
print(df1)
print(df4)
sonuc = pd.concat([df1, df4], axis=1, join="outer") #default outer. veri kaybı olmaz.
sonuc
```

```python
sonuc = pd.concat([df1, df4], axis=1, join="inner") #ikisinde ortak indekslileri birbirine ekler
sonuc
```
